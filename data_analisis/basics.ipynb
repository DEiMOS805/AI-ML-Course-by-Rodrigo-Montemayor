```python
data_na = data.dropna()
data_na.info()
```

```python
# Map the categorical variable to numerical values using dummies or one-hot encoding
# NEAR BAY INLAND NEAR OCEAN
# 1        0      0
# 0        1      0
# 0        0      1

dummies = pd.get_dummies(data_na["ocean_proximity"], dtype=int)
data_na = pd.concat([data_na, dummies], axis=1) # data_na.join(dummies)
```

```python
data_na
```

```python
data_na = data_na.drop(columns=["ocean_proximity"])
data_na
```
