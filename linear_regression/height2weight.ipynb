```python
# for data visualization
import seaborn as sb
import matplotlib.pyplot as plt

# for data manipulation
import numpy as np
import pandas as pd
from pandas import DataFrame


# for data analysis
from sklearn.linear_model import LinearRegression
```

```python
# Read the data
df: DataFrame = pd.read_csv("./height2weight.csv")
```

```python
sb.scatterplot(
	data=df, x="height", y="weight",
	hue="weight",
	palette=sb.diverging_palette(250, 30, l=65, center="dark", as_cmap=True)
)
```

```python
plt.figure(figsize=(8, 6))
sb.regplot(
	data=df, x="height", y="weight",
	scatter_kws={"s": 50, "alpha": 0.7},
	line_kws={"color": "red", "linewidth": 1}
)
plt.title("Weight vs Height")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.grid(True)
plt.show()
```

```python
x = df["height"]
y = df["weight"]
```

```python
x_reshape = x.values.reshape(-1, 1)
y_reshape = y.values.reshape(-1, 1)
```

```python
model = LinearRegression()
```

```python
model.fit(x_reshape, y_reshape)
```

```python
height_pred = np.arange(1.4, 3.0, 0.2).reshape(-1, 1)
weight_pred = model.predict(height_pred)
height_pred
```

```python
# Mostrar los datos y la línea de regresión
plt.scatter(x_reshape, y_reshape, color='blue', label='Datos reales')
plt.plot(height_pred, weight_pred, color='red', label='Línea de regresión')
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")
plt.legend()

plt.xlim([1.4, 2.0])
plt.ylim([0, 100])


plt.grid(True)
plt.show()
```

```python
model.score(x_reshape, y_reshape)
```
